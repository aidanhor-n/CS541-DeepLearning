```python
def mean_of_feature(dataframe, feature):
    powders = pd.DataFrame(dataframe)
    all_vals = []
    for i in powders[feature]:
        all_vals.append(i)
    return(np.mean(all_vals))
```

```python
def std_of_feature(dataframe, feature):
    powders = pd.DataFrame(dataframe)
    all_vals = []
    for i in powders[feature]:
        all_vals.append(i)
    return(np.std(all_vals))
```

```python
def larger_std(dataframe, feature):
    powders = pd.DataFrame(dataframe)
    std = std_of_feature(powders, feature)
    avg = mean_of_feature(powders, feature)
    if(std > avg):
        return(0)
    else:
        return(1)
```

```python
def std_mean_ratio(dataframe, feature):
    powders = pd.DataFrame(dataframe)
    std = std_of_feature(powders, feature)
    avg = mean_of_feature(powders, feature)
    std_ratio = std/avg
    return(std_ratio)
```

```python
def graph_feature(dataframe, feature, bin_size):
    powders = pd.DataFrame(dataframe)
    all_vals = []
    for i in powders[feature]:
        all_vals.append(i)
    plt.hist(all_vals, bins=bin_size)
    plt.xlabel(str(feature))
    plt.ylabel('Occurrences')
    plt.show()
```

```python
def show_mean_std_all(data, features): 
    for i in features:
        print(i)
        print(mean_of_feature(data, i))
        print(std_of_feature(data, i))
        print("="*25)
```

```python
def show_bad_features(data, features):
    bad_features = []
    for i in features:
        if larger_std(data, i) == 0:
            bad_features.append(i)
            print(i)
            print('Mean: ' + str(mean_of_feature(data, i)) + ' - Std: ' + str(std_of_feature(data, i)))
            print('Ratio (std:mean) - ' + str(std_mean_ratio(data, i)) + ':1')
            print('='*25)
    print(bad_features)
        
```

```python
def graph_all(data, features, bin_size):
    for i in features:
        print(i)
        graph_feature(data, i, bin_size)
        print('='*25)
```

```python
def count_zero(data, feature):
    num_zero = 0
    zero_frame = pd.DataFrame(data)
    for i in zero_frame[feature]:
        if i == 0:
            num_zero+=1
    return(num_zero)
```

```python
def show_all_zeros(data, features):
    for i in features:
        zeros = count_zero(data, i)
        print(i + ' has ' + str(zeros) + " 0's in it's data" + '\n' + '='*40)
```

```python
def print_all_vals(data, feature):
    vals = []
    powder = pd.DataFrame(data)
    for i in powder[feature]:
        vals.append(i)
    vals.sort()
    return(vals)
```

```python
def remove_weight(data, feature, max):
    vals = []
    powder = pd.DataFrame(data)
    for i in powder[feature]:
        vals.append(i)
    vals.sort()
    deweighted_vals = (max*(vals - np.min(vals))/np.ptp(vals)).astype(float)
    return(deweighted_vals)
```

```python
def analyze_all(data, bin_size):
    powders = pd.DataFrame(data)
    for i in powders:
        if i != "FlowClass" and i != "Flow":
            print('Feature: ' + i)
            print('Mean: ' + str(mean_of_feature(data, i)))
            print('Std: ' + str(std_of_feature(data, i)))
            print('Std:Mean Ratio: ' + str(std_mean_ratio(data, i)))
            graph_feature(data, i, bin_size)
            print('Number of 0\'s: ' + str(count_zero(data, i)))
            print('='*60)
```
