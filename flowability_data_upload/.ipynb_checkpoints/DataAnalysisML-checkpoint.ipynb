```python
def compute_correlation(df, data_folder, method, threshold, heldout_cols = None):
        #x_cols = [col for col in df.columns if col not in ['SampleName', 'Flow', 'FlowClass']]
        x_cols = [col for col in df.columns if col not in []]
        if heldout_cols:
            for col in heldout_cols:
                x_cols.remove(col)
        x_data = df[x_cols]
        #files = os.listdir('.')
        if method == 'pearson':
            # linear correlation
            corr_matrix = x_data.corr('pearson')
        else: 
            if not os.path.exists(f'{data_folder}/nonlinear_correlation_matrix.pkl'):
                # nonlinear correlation
                # subset = np.arange(2000)
                # np.random.shuffle(subset)
                print("Recomputing MIC matrix.")
                x_data = x_data.sample(n=2000, replace = False)
                mine = MINE(alpha=0.6, c=15, est="mic_approx")
                corr_matrix = [[0]*len(x_cols) for i in range(len(x_cols))]
                for i in range(len(x_cols)):
                    i_col = x_cols[i]
                    for j in range(len(x_cols)):
                        if i == j:
                            corr_matrix[i][j] = 1
                        else: 
                            j_col = x_cols[j]
                            i_data = x_data[i_col]
                            j_data = x_data[j_col]
                            mine.compute_score(i_data, j_data)
                            corr_matrix[i][j] = abs(mine.mic())
                
                corr_matrix = np.asarray(corr_matrix)
                with open(f'{data_folder}/nonlinear_correlation_matrix.pkl', 'wb') as f:
                    pkl.dump(corr_matrix, f)
            else:
                with open(f'{data_folder}/nonlinear_correlation_matrix.pkl', 'rb') as f:
                    corr_matrix = pkl.load(f)
        for x in corr_matrix:
            corr_matrix[x] = abs(corr_matrix[x])
        return corr_matrix
```

```python
def visualize_correlation(self, output_file, method='pearson', threshold=0.8):
        df = self
        files = os.listdir('.')
        method_names = {'pearson': "Pearson", "mic": "MIC"}
        f, ax = plt.subplots(figsize=(11, 9))
        corr_matrix = compute_correlation(df, "n", method, threshold)
        #x_cols = [col for col in df.columns if col not in ['SampleName', 'Flow', 'FlowClass']]
        x_cols = [col for col in df.columns if col not in []]


        sns.heatmap(corr_matrix, xticklabels=x_cols, yticklabels=x_cols)
        
        plt.title("{} Correlation for Full Data Matrix".format(method_names[method]))
        plt.savefig(output_file)
        #output_file.close()
```
